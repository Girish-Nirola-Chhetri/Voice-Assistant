```python
import json

with open("../data/raw/alarm_data.json", "r", encoding="utf-8") as f:
    data = json.load(f)


with open("sp_training_data.txt", "w", encoding="utf-8") as f:
    for item in data:
        f.write( item["command"] + "\n")
        f.write( json.dumps( item["json"]) + "\n\n")
```

```python
import sentencepiece as spm

spm.SentencePieceTrainer.train(
    input="sp_training_data.txt",
    model_prefix="assistant_tokenizer",
    vocab_size=200,
    model_type="bpe"
)
```

```python
# testing the tokenizer

tokenizer = spm.SentencePieceProcessor()
tokenizer.Load("assistant_tokenizer.model")


print(tokenizer.encode_as_pieces(
    "Set an alarm for 6 AM"
))

print(tokenizer.encode_as_ids(
    "Set an alarm for 6 AM"
))
```
